```python
import numpy as np
import matplotlib.pyplot as plt
import random

from tqdm.auto import tqdm

from simulator import *
```

```python
femto_file_irm = f"./figures/pickle/femto_irm_caches.pickle"
femto_file_rrm = f"./figures/pickle/femto_rrm_caches.pickle"
femto_file_yt = f"./figures/pickle/femto_yt_caches.pickle"
femto_file_movielens = f"./figures/pickle/femto_movielens_caches.pickle"
```

# IRM

```python
(cache_DEC, cache_DEC_mix, cache_BSA, cache_D_BSA, cache_mLRU, cache_LazyLRU, cache_BH) = load_caches(femto_file_irm)
cache_mLRU.name = "mLRU"
cache_LazyLRU.name = "LazyLRU"
cache_BSA.name = "BSA"
cache_D_BSA.name = "D-BSA"
cache_DEC.name = "MultiDEC"
cache_DEC_mix.name = "MultiDEC (Mix)"
cache_BH.name = "~BH"
```

```python
dataset = "irm"
save = True


###########################
set_fig_fullwidth(scale=1.5)
plot_comp(
    cache_mLRU,
    cache_LazyLRU,
    cache_BH,
    cache_BSA,
#     cache_D_BSA,
    cache_DEC,
#     cache_DEC_mix,
    ylabel="Utility",
    legend_columns=2
)
if save:
    plt.savefig(f"./figures/femto_{dataset}.pdf")
plt.show()
###########################
set_fig_onethirdwidth()
for i, expert_cache in enumerate(cache_DEC.caches):
    plot_expert_choices(expert_cache, ["LRU", "FTPL", "D-OGA"])
    if save:
        plt.savefig(f"./figures/femto_{dataset}_ep_choices_{i+1}.png", dpi=600)
    
    plt.show()

###########################
set_fig_fullwidth(scale=1.5)
plot_comp(
    cache_mLRU,
    cache_LazyLRU,
    cache_BH,
    cache_BSA,
    cache_D_BSA,
    cache_DEC,
    cache_DEC_mix,
    ylabel="Utility",
    legend_columns=2
)

if save:
    plt.savefig(f"./figures/femto_{dataset}_all.pdf")
plt.show()
###########################
set_fig_halfwidth()
plot_comp(
    cache_BSA,
    cache_D_BSA,
    ylabel="Utility",
)

if save:
    plt.savefig(f"./figures/D-BSA_{dataset}.pdf")
plt.show()
    
    
    
```
